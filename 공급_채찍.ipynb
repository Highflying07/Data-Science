```python
import numpy as np
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(42)

# Simulation settings
time_steps = 30
base_demand = 100           # Average customer demand
demand_fluctuation = 10     # Random fluctuation range

# Amplification factors (how much each stage overreacts)
retailer_amplification = 1.2
wholesaler_amplification = 1.5
manufacturer_amplification = 1.8

# Generate random customer demand
customer_demand = base_demand + np.random.randint(-demand_fluctuation, demand_fluctuation, time_steps)

# Initialize orders at each stage
retailer_orders = []
wholesaler_orders = []
manufacturer_orders = []

# Set initial orders
retailer_orders.append(customer_demand[0] * retailer_amplification)
wholesaler_orders.append(retailer_orders[0] * wholesaler_amplification)
manufacturer_orders.append(wholesaler_orders[0] * manufacturer_amplification)

# Run simulation over time
for t in range(1, time_steps):
    r_order = customer_demand[t] * retailer_amplification
    w_order = r_order * wholesaler_amplification
    m_order = w_order * manufacturer_amplification
    retailer_orders.append(r_order)
    wholesaler_orders.append(w_order)
    manufacturer_orders.append(m_order)

# Plot results
plt.figure(figsize=(12, 6))
plt.plot(customer_demand, label='Customer Demand', color='blue', linewidth=2)
plt.plot(retailer_orders, label='Retailer Orders', color='green', linestyle='--')
plt.plot(wholesaler_orders, label='Wholesaler Orders', color='orange', linestyle='--')
plt.plot(manufacturer_orders, label='Manufacturer Orders', color='red', linestyle='--')
plt.xlabel('Time')
plt.ylabel('Quantity')
plt.title('Bullwhip Effect Simulation (No Improvements)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Improvement strategies (lower amplification by better management)
retailer_amplification_improved = 1.05   # Reduced amplification with POS data sharing
wholesaler_amplification_improved = 1.1  # Smoothed orders with fixed schedule
manufacturer_amplification_improved = 1.2 # Better forecasting with AI

# Initialize improved orders
retailer_orders_improved = []
wholesaler_orders_improved = []
manufacturer_orders_improved = []

# Set initial improved orders
retailer_orders_improved.append(customer_demand[0] * retailer_amplification_improved)
wholesaler_orders_improved.append(retailer_orders_improved[0] * wholesaler_amplification_improved)
manufacturer_orders_improved.append(wholesaler_orders_improved[0] * manufacturer_amplification_improved)

# Run improved simulation over time
for t in range(1, time_steps):
    r_order_imp = customer_demand[t] * retailer_amplification_improved
    w_order_imp = r_order_imp * wholesaler_amplification_improved
    m_order_imp = w_order_imp * manufacturer_amplification_improved
    retailer_orders_improved.append(r_order_imp)
    wholesaler_orders_improved.append(w_order_imp)
    manufacturer_orders_improved.append(m_order_imp)

# Plot improved results
plt.figure(figsize=(12, 6))
plt.plot(customer_demand, label='Customer Demand', color='blue', linewidth=2)
plt.plot(retailer_orders_improved, label='Retailer Orders (Improved)', color='green', linestyle='--')
plt.plot(wholesaler_orders_improved, label='Wholesaler Orders (Improved)', color='orange', linestyle='--')
plt.plot(manufacturer_orders_improved, label='Manufacturer Orders (Improved)', color='red', linestyle='--')
plt.xlabel('Time')
plt.ylabel('Quantity')
plt.title('Bullwhip Effect Simulation with Improvements')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Initial settings
np.random.seed(42)           # Set random seed for reproducibility (so results are consistent)
timesteps = 30               # Number of time steps to simulate
initial_anxiety = 1          # Initial anxiety level

# Amplification rate for bullwhip effect
amplification_rate = 1.2     # Amplification factor (how much anxiety increases over time)

# Reduction factors for improvement strategies
info_sharing_effect = 0.7    # Effect of information sharing strategy
routine_effect = 0.8         # Effect of routine/habit strategy
self_awareness_effect = 0.6  # Effect of self-awareness or mindfulness strategy

# Simulation function
def simulate_anxiety(amplification, reduction_effect=1.0):
    anxiety = [initial_anxiety]  # Start with initial anxiety level in a list
    for t in range(1, timesteps):
        # Calculate new anxiety:
        # multiply last anxiety by amplification and reduction factor,
        # then add small random noise from normal distribution
        new_anxiety = anxiety[-1] * amplification * reduction_effect + np.random.normal(0, 0.2)
        # Avoid negative anxiety (set minimum to zero)
        anxiety.append(max(new_anxiety, 0))
    return anxiety  # Return the full anxiety time series

# Run simulations for four scenarios
no_strategy = simulate_anxiety(amplification_rate, reduction_effect=1.0)                    # No intervention
with_info_sharing = simulate_anxiety(amplification_rate, reduction_effect=info_sharing_effect)  # With information sharing
with_routine = simulate_anxiety(amplification_rate, reduction_effect=routine_effect)           # With routine practice
with_self_awareness = simulate_anxiety(amplification_rate, reduction_effect=self_awareness_effect)  # With self-awareness

# Plotting the results
plt.figure(figsize=(12, 6))  # Set the size of the figure
plt.plot(no_strategy, label='Anxiety Amplification (No Intervention)', color='red', linewidth=2)
# Plot the anxiety line with no intervention (red, solid line)
# plt.plot(with_info_sharing, label='Information Sharing Effect', color='green', linestyle='-')
# # Plot the anxiety line with information sharing (green, dashed line)
# plt.plot(with_routine, label='Routine Effect', color='blue', linestyle='-.')
# # Plot the anxiety line with routine/habit strategy (blue, dashed line)
plt.plot(with_self_awareness, label='Self-Awareness Effect', color='purple', linestyle=':')

# Plot the anxiety line with self-awareness/mindfulness (purple, dashed line)
plt.xlabel('Time (Days, Weeks, etc.)')  # Label for x-axis
plt.ylabel('Anxiety Level')             # Label for y-axis
plt.title('Simulation of Inner Anxiety Bullwhip Effect')  # Title of the chart
plt.legend()                            # Show legend to explain lines
plt.grid(True)                          # Add grid for easier reading
plt.show()                              # Display the plot
```
